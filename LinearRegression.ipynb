```python
import numpy as np
import pandas as pd
```

```python
class LinearRegression:
    def __init__(self,num_iters,alpha):
        self.num_iters = num_iters
        self.alpha = alpha
        
    def loss(self,y,preds):
        return np.sum(np.square(y-preds))/y.shape[0]
        
    def fit(self,X,y):
        
        self.weights = np.zeros(X.shape[1])
        self.bias = 0
        
        preds = self.predict(X)
        
        current_loss =  self.loss(y,preds)
        
        for epoch in range(self.num_iters):
            
            #Compute the gradients w.r.t weights and bias
            dw = (2/y.shape[0]) * np.dot(X.T,(preds-y))
            
            db = (2/y.shape[0]) * np.sum(y-preds)
            
            #update the weights and bias
            self.weights -= dw * self.alpha
            self.bias -= db * self.alpha
            
#             print(self.weights)
            
            #make predictions
            preds = self.predict(X)
            
            #calculate new loss
            new_loss = self.loss(y,preds)

            print(new_loss)
            if current_loss < new_loss:
                break
            
            current_loss = new_loss         
            
            
        
    
    def predict(self,X):
        z = np.dot(X,self.weights) + self.bias
        
        return z
        
        
```

```python
import sklearn
from sklearn import datasets
X, y = datasets.make_regression(n_samples=100,n_features=1,noise=20)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],y,color='b')
```

```python
model = LinearRegression(100000,0.001)
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
laass = model.loss(y_test,predictions)
```
